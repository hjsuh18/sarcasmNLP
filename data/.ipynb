This is a notebook to try out importing and saving the SARC dataset using Python. The ultimate aim is to write a Python file that processes and saves the data to DynamoDB.

```python
import ijson
# import json
# import pandas as pd
from pprint import pprint
```

```python
f = open('sarc/main/comments.json', 'r')
```

```python
parser = ijson.parse(f)
```

```python
parser
```

```python
# go through json file
for prefix, event, value in parser:
    if event == 'map_key':
        k = value
        v = {}
        while True:
            prefix, event, value = next(parser)
            if event == 'end_map':
                break
            if event == 'map_key':
                key = value
                prefix, event, value = next(parser)
                v[key] = value
        
        print("key: " + k)
        print("value: ")
        for a, b in v.items():
            print(str(a) + ":" + str(type(b)))
```
